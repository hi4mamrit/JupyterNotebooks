# a REST API to find links of recently unanswered python questions in stackOverflow

```python
"""
visit https://api.stackexchange.com/  for various apis of stackoverflow (SO)

e.g of 1 api under questions section:- /2.2/questions?order=desc&sort=activity&site=stackoverflow
The complete api will be:-
    http://api.stackexchange.com/2.2/questions?order=desc&sort=activity&site=stackoverflow

curl way to POST a data to api:-
in windows:-
  curl -X POST -H "Content-Type: application/json" -d "{ \"key1\": \"value1\" }" http://localhost:5000/api/method
or use POSTMAN app
"""

print("Idea to start with -->")
```

```python
#consuming the api
import requests
response = requests.get("http://api.stackexchange.com/2.2/questions?order=desc&sort=activity&site=stackoverflow")
```

```python
#Converting response to json and examining the structure
response.json()
```

```python
# Interest in items key
response_items = response.json()['items']
response_items  # list of dictionaries
```

```python
print(response_items[0].keys() ," <- All the keys of a question in StackOverflow")
```

```python
#is_answered(:boolean) & link(:str) & tags(:list)  <- these are the variables I need , Let's program a function

def f_na_python(p_list):
    #Takes list as input , iterates over it filtering 'interested' tags and resturns the list of links as result
    v_list= list()
    
    for item in p_list:
        if not bool(item['is_answered']) and ('python' in [x.lower() for x in item['tags']]):
            v_list.append(item['link'])
    
    return v_list
```

```python
links = f_na_python(response_items)
print(*links,sep='\n')
```

```python
#Let's make the function more interesting and generic to find 'unanswered' questions related to any language

def f_na_language(p_list,p_language):
    #Takes list as input , iterates over it filtering 'interested' tags and resturns the list of links as result
    v_list= list()
    v_language = p_language.lower()
    
    for item in p_list:
        if not bool(item['is_answered']) and (v_language in [x.lower() for x in item['tags']]):
            v_list.append(item['link'])
    
    return v_list
```

```python
python_links = f_na_language(response_items,'Python')
print('Python Un-answered links')
print(*python_links,sep='\n')

java_links = f_na_language(response_items,'Java')
print('Java Un-answered links')
print(*java_links,sep='\n')
```
